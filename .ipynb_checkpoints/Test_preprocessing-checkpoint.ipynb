```python
import category_encoders as ce
import pandas as pd
```

```python
df_train = pd.read_csv("../Data/Final/train_data.csv")
df_test = pd.read_csv("../Data/Raw/test.csv")
```

```python
df_train
```

```python
df_test
```

```python
df_test = df_test.iloc[::-1]
df_test['ofd_date'] = pd.to_datetime(df_test['ofd_date'], format = "%Y-%m-%d")
df_test["weekday"] = df_test["ofd_date"].dt.weekday
df_test["week_no"] = df_test["ofd_date"].dt.isocalendar().week
```

```python
df_test_MNR = df_test.copy()
df_test_Earl = df_test.copy()
```

```python
target_encoder = ce.TargetEncoder(cols=["country_code", "station_code"], return_df = True, verbose=1)
```

```python
df_temp = df_train.loc[:,["country_code", "station_code"]]
target_encoder.fit(X = df_temp, y = df_train["MNR_SNR_Exp"])
df_temp = df_test.loc[:,["country_code", "station_code"]]
df_test_MNR.drop(df_test_MNR.columns[[1,3]], axis=1, inplace=True)
df_test_MNR = pd.concat([df_test_MNR, target_encoder.transform(X = df_temp)], axis=1)
df_temp = df_train.loc[:,["country_code", "station_code"]]
target_encoder.fit(X = df_temp, y = df_train["Earlies_Exp"])
df_temp = df_test.loc[:,["country_code", "station_code"]]
df_test_Earl.drop(df_test_Earl.columns[[1,3]], axis=1, inplace=True)
df_test_Earl = pd.concat([df_test_Earl, target_encoder.transform(X = df_temp)], axis=1)
```

```python
binary_encoder = ce.BinaryEncoder(cols=["fc_codes"], return_df = True, verbose=1)
```

```python
df_temp = df_train.loc[:,["fc_codes"]]
binary_encoder.fit(X = df_temp, y = df_train["MNR_SNR_Exp"])
df_temp = df_test.loc[:,["fc_codes"]]
df_test_MNR.drop(df_test_MNR.columns[1], axis=1, inplace=True)
df_test_MNR = pd.concat([df_test_MNR, binary_encoder.transform(X = df_temp)], axis=1)
df_temp = df_train.loc[:,["fc_codes"]]
binary_encoder.fit(X = df_temp, y = df_train["Earlies_Exp"])
df_temp = df_test.loc[:,["fc_codes"]]
df_test_Earl.drop(df_test_Earl.columns[1], axis=1, inplace=True)
df_test_Earl = pd.concat([df_test_Earl, binary_encoder.transform(X = df_temp)], axis=1)
```
