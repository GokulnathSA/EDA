## Outlier removal method

 - Separate our dataframe based on the delivery centre. The columns to remove outliers for are
     - OFD
     - Slam
     - Earlies_Rec
     - Rollover
     - Returns
     - R_Sideline
     - Sideline
     - Earlies_Exp
     - MNR_SNR_Exp
 - Compare the OFD actual value vs the formula value to see how big the [other] value is
 - Out of these, OFD depends on other columns listed and Earlier_Exp and MNR_SNR_Exp are targets, don't touch them directly. Try and apply STL decomposition column wise to reduce the residual amounts on all the other columns using the IQR method and then recombine them to form the final columns
 - Delete rows based on outliers in Earlier_Exp and MNR_SNR_Exp using IQR method and finally recalculate the value of OFD using the new column values

```python
df["OFD_Expected"] = df["Slam"] + df["Earlies_Exp"] - df["Earlies_Rec"] - df["MNR_SNR_Exp"] + df["Rollover"] + df["Returns"] + df["R_Sideline"] - df["Sideline"]
df.plot(x="ofd_date", y=["OFD","OFD_Expected"], figsize=(20,8))
```

```python
plt.plot(df["ofd_date"], df["OFD"] - df["OFD_Expected"])
plt.gcf().set_size_inches((20,8))
```

```python
df["other"] = df["OFD"] - df["OFD_Expected"]
```

```python
df_dict = {}
delivery_centres = [ f"Delivery_D{i+1}" for i in range(80) ]
for centre in delivery_centres:
    df_dict[centre] = df[df[centre] == 1].copy()
```

```python
df_dict["Delivery_D1"]
```

```python
result = STL(df_dict["Delivery_D1"]["Slam"], period=2).fit()
result.plot()
plt.gcf().set_size_inches((20,8))
plt.show()
```

```python
result = STL(df_dict["Delivery_D1"]["Earlies_Rec"], period=2).fit()
result.plot()
plt.gcf().set_size_inches((20,8))
plt.show()
```

```python
columns = {"Slam":2, "Earlies_Rec":2, "Rollover":2, "Returns":8, "R_Sideline":8, "Sideline":8}
for centre in delivery_centres:
    for col, period in columns.items():
        result = STL(df_dict[centre][col], period = period, seasonal_deg=1).fit()
        Q3 = np.percentile(result.resid, 75)
        Q1 = np.percentile(result.resid, 25)
        IQR = Q3-Q1
        upper = Q3 + (1.5 * IQR)
        mask = (result.resid > upper)
        result.resid[mask] = np.median(result.resid)
        df_dict[centre][col] = (result.seasonal+result.trend+result.resid).round()
```

```python
df = pd.concat(df_dict.values())
```

```python
df.sort_index(inplace=True)
```
