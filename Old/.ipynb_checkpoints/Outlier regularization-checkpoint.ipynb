```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
```

```python
df["OFD_Expected"] = df["Slam"] + df["Earlies_Exp"] - df["Earlies_Rec"] - df["MNR_SNR_Exp"] + df["Rollover"] + df["Returns"] + df["R_Sideline"] - df["Sideline"]
```

```python
def outlier_point_highlight(col, outlier_indices):
    outliers = np.empty(len(col))
    outliers[:] = np.nan
    for i,outlier_index in enumerate(outlier_indices):
        if outlier_index == -1:
            outliers[i] = col[i]
    return outliers
```

```python
df_dict = dict()
delivery_cols = [ f"Delivery_D{i+1}" for i in range(80) ]
for del_centre in delivery_cols:
    df_dict[del_centre] = df[df[del_centre] == 1].copy()
```

```python
df[df["Delivery_D1"] == 1]
```

```python
for del_centre in delivery_cols:
    for col in df.columns[4:12]:
        outlier_indices = IsolationForest(contamination=0.03, random_state=42).fit_predict(df_dict[del_centre][col].values.reshape(-1,1))
        outliers = outlier_point_highlight(df_dict[del_centre][col].values, outlier_indices)
        df_dict[del_centre][col] = df_dict[del_centre][col].where(~df_dict[del_centre][col].isin(outliers), np.nan)
```

```python
for del_centre in delivery_cols:
    df_dict[del_centre].interpolate(method='ffill', inplace=True)
for del_centre in delivery_cols:
    df_dict[del_centre].interpolate(method='bfill', inplace=True)
```

```python
df = pd.concat(df_dict.values())
```

```python
df = df.reset_index().drop(labels = 'index', axis=1)
```

```python
df
```

```python
max(df.isna().sum())
```

```python
df["OFD_Expected"] = df["Slam"] + df["Earlies_Exp"] - df["Earlies_Rec"] - df["MNR_SNR_Exp"] + df["Rollover"] + df["Returns"] + df["R_Sideline"] - df["Sideline"]
df["other"] = df["OFD"] - df["OFD_Expected"]
```

```python
df.to_excel("..\Data\Final\\train_data_revised.xlsx", index=False)
```
