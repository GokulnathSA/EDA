```python
import sklearn
from sklearn.ensemble import IsolationForest
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sklearn.__version__
```

```python
df = pd.read_excel("..\Data\Preprocessed\\train_data.xlsx")
```

```python
df.info()
```

```python
def outlier_point_highlight(OFD, outlier_indices):
    outliers = np.empty(len(OFD))
    outliers[:] = np.nan
    for i,outlier_index in enumerate(outlier_indices):
        if outlier_index == -1:
            outliers[i] = OFD[i]
    return outliers
```

```python
df["OFD"].values
```

```python
outlier_indices = IsolationForest(contamination=0.03).fit_predict(df["OFD"].values.reshape(-1,1))

outliers = outlier_point_highlight(df["OFD"].values, outlier_indices)

plt.plot(df["ofd_date"], df["OFD"])
plt.plot(df["ofd_date"], outliers, "r+")
plt.gcf().set_size_inches((20,8))
```

```python
plt.scatter(df["ofd_date"], outlier_indices)
```

```python
plt.scatter(df["ofd_date"], ~df["OFD"].isin(outliers))
```
