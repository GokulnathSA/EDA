```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
```

```python
df = pd.read_excel("..\Data\Preprocessed\\train_data.xlsx")
```

```python
df.info()
```

```python
def outlier_point_highlight(cols, outlier_indices):
    cols = cols.transpose()
    outlier_indices = outlier_indices.transpose()
    outliers = np.empty(cols.shape)
    outliers[:,:] = np.nan
    for i,col in enumerate(cols):
        for j,outlier_index in enumerate(outlier_indices):
            if outlier_index == -1:
                outliers[i,j] = cols[i,j]
    return outliers
```

```python
cols = df.columns[2:10]
outlier_indices = IsolationForest(contamination=0.03).fit_predict(df[cols].values)
outliers = outlier_point_highlight(df[cols].values, outlier_indices)
for (outlier, col) in zip(outliers,cols):
    df[col] = df[col].where(~df[col].isin(outlier), np.nan)
```

```python
df = df.dropna().reset_index()
```

As we can see, trying to remove all outliers simultaneously from the entire dataframe using this method removes far too many elements, hence the previous method was most appropriate
